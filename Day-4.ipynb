### Separate the input and output variables 

```python
x = df[['disp', 'hp', 'wt']]
y = df['mpg']
```

```python
x
```

### Import the class

```python
from sklearn.linear_model import LinearRegression
```

### Create the object

```python
reg = LinearRegression()
```

### Train the Model

```python
reg.fit(x,y)
```

### Predict on unseen data

```python
disp = 221
hp = 102
wt = 2.91
reg.predict([[disp, hp, wt]])
```

### Visualize

```python
plt.figure(figsize=(16, 9))
plt.subplot(2,2,1)
plt.scatter(df['disp'], y, color='r')
plt.subplot(2,2,2)
plt.scatter(df['hp'], y, color='b')
plt.subplot(2,2,3)
plt.scatter(df['wt'], y, color='g')
```

```python
reg.score(x,y)
```

# Decision Tree Classification

### Import the dataset

```python
df = pd.read_csv('Social_Network_Ads.csv')
```

```python
df.shape
```

```python
list(df.columns)
```

```python
df
```

### Saparate the input and output variables

```python
x = df[['Age', 'EstimatedSalary']]
y = df['Purchased']
```

```python
x
```

### Explore the data

```python
plt.xlabel('Age')
plt.ylabel('Salary')
plt.grid()
plt.scatter(x['Age'], x['EstimatedSalary'], c=y)
```

### Import the class

```python
from sklearn.tree import DecisionTreeClassifier
```

### Create the object

```python
classifier = DecisionTreeClassifier(random_state=0)
```

### Train the Model

```python
classifier.fit(x, y)
```

### Plot the tree

```python
from sklearn.tree import plot_tree
import matplotlib.pyplot as plt
```

```python
plt.figure(figsize=(12,15))
plot_tree(classifier, fontsize=10, feature_names=['age', 'sal'],
         class_names=['y','n'], filled=True, rounded=True);
```

### Prediction on unseen data

```python
new = pd.DataFrame({
    'Age': [25,37,56,66],
    'EstimatedSalary': [34000,41000,26000,134000]
})
```

```python
new
```

```python
classifier.predict(new)
```
